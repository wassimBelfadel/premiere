# Etre organisé est FONDAMENTAL en informatique 


## Comment nommer son TP ?

**Vous n'avez pas le choix du nom de votre document!!**. 

Prenons un exemple : lors d'une séance je vous demande de faire un TP appelé `TP_machinTruc`. Imaginons que l'énoncé de ce TP se trouve dans `premiere/bloc6/sousDossier2/sousSousDossier7/`. Imaginons que pour faire ce TP vous allez devoir écrire un programme python. Comment procéder ?

1. Vous ouvrez Thonny pour écrire votre programme
2. **La première chose à faire AVANT de commencer à coder est de sauvegarder ce document** (même s'il est encore vierge)
3. Pour la sauvegarde, Thonny va vous demander un nom de fichier. Vous l'appelerez **obligatoirement** `TP_machinTruc` en respectant [la casse](https://fr.wikipedia.org/wiki/Sensibilit%C3%A9_%C3%A0_la_casse) et vous le sauvegarderez **obligatoirement dans `/home`**
4. Inutile de rajouter l'extension `.py`. Thonny va la rajouter automatiquement.

Imaginons que pour faire ce TP, vous devez aussi écrire un compte rendu de TP : 
1. Vous le rédigerez en **Markdown** (voir document explicatif)
2. Vous l'appelerez aussi `TP_machinTruc`. Mais l'extension sera différente. Ce sera `.md` 

Imaginons que pour ce même TP, vous devez écrire un autre fichier, il s'appelera aussi `TP_machinTruc` mais avec une autre extension (qui va dépendre du logiciel utilisé)


## Sauvegarder son travail

> Lorsque vous faites un TP, il peut se produire un _"plantage"_ du logiciel, de la machine, une coupure de courant, ou tout autre mauvaise manipulation à nimporte quel moment. Dans ce cas, **tout travail non sauvegardé est perdu, souvent définitivement**. Il arrive trop souvent que des élèves perdent leur travail parce qu'ils ne sauvegardent pas leur TP avant la fin de la séance.

* Si vous perdez votre travail à cause d'un problème, même un problème indépendant de votre volonté, ça sera **toujours VOTRE faute**. Si le travail est noté : tant pis pour vous ! Le travail est non rendu donc vous aurez 0/20.

* Pour éviter cela lorsque vous êtes en TP, vous devez sauvegarder **TRES REGULIEREMENT** votre travail, par exemple toutes les 10 minutes. Ainsi le pire qu'il puisse vous arrivez c'est perdre 10 minutes de travail

**Comment savoir si mon travail est sauvegardé ?**

1. Un document non sauvegardé s'appelle souvent **Untitled** par défaut.
2. Lorsque vous modifiez votre document, il apparaît souvent une petite étoile ou un changement de couleur de l'icône de sauvegarde. Cela signifie que vos dernières modifications n'ont pas été sauvegardées

**Où sauvegarder son travail ?**

* La sauvegarde **temporaire** doit se faire dans le répertoire `/home`. **RAPPEL : Tout document laissé dans `/home` est susceptible d'être supprimé en fin de séance après votre départ de la salle**. La sauvegarde est dont bien **temporaire**, juste le temps de la séance.
* La sauvegarde **définitive** doit se faire **en fin de séance ou de TP** dans votre _repository_ **travail** en respectant **l'arborescence** (voir ci-dessous)
 
## Comment organiser son _repository_ travail dans github ?

* Il suffit donc de respecter l'**arborescence** de mon _repository_ **premiere** dans votre  _repository_ **travail**

* Reprenons l'exemple du `TP_machinTruc` dont l'énoncé se trouve dans `premiere/bloc6/sousDossier2/sousSousDossier7/`

votre ou vos fichiers (qui s'appellent donc tous `TP_machinTruc`) doivent être rangé dans `travail/bloc6/sousDossier2/sousSousDossier7/`


## Pour Conclure

1. Il arrive souvent qu'un programme python ou qu'un site web a un problème uniquement à cause d'un problème de nommage ou d'arborescence. Ces erreurs sont souvent délicates à trouver parce que certains élèves font vraiment des trucs tordus sans s'en rendre compte dans la pagaille de leurs fichiers ou dossiers.
2. J'ai au total une petite trentaine d'élève en NSI en tout (première et terminale). Je ne vais pas pour chaque élève, _"m'amuser"_ à chercher où vous avez bien pû mettre votre TP parce que vous n'avez pas respecté l'arborescence ni ouvrir plein de fichiers parce que vous n'avez pas respecté le nom du fichier.

Bref, les problèmes de nom et d'arborescence me font perdre beaucoup de temps. En conséquence :

### Si vous perdez votre travail car il n'a pas été sauvegardé régulièrement, si vous ne nommez correctement vos fichiers ou si vous ne respectez pas l'arborescence : VOUS AUREZ 0/20 au TP!
