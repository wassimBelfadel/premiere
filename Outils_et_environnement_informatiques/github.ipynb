# Création de votre repository privé personnel sur github

Dans un premier temps, vous verrez **github** comme une espèce de _drive_ hébergé dans le _cloud_ (un peu comme googleDrive ou DropBox). En fait c'est bien plus que cela et vous le constaterez progressivement en l'utilisant en cours d'année. **github** est un outil professionnel utilisé par de très nombreux informaticiens dans le développement de leurs projets informatiques.

Dans le jargon, https://github.com/seynave/premiere s'appelle un _repository_. Celui-ci est **public**, c'est pour cette raison que vous, comme nimporte quel utilisateur, avez pû y accéder sans vous identifier : la **lecture** est ouverte à tous, mais pas l'**écriture**.

Vous allez maintenant créer un _repository_ **privé** dont l'accès est sécurisé par mot de passe pour y stocker **votre** travail fait en NSI. Ainsi ce travail sera accessible uniquement par vous (et par moi) depuis la classe mais aussi depuis nimporte quelle machine (ordinateur ou smartphone) en dehors de la classe. C'est le **seul endroit où sauvegarder votre travail** (il est cependant vivement conseillé d'en faire des sauvegardes régulièrement chez vous)

>**NE PAS SE PRECIPITER : LIRE TOUTE LA SUITE JUSQU'A ARRIVER AU POINT 2 AVANT DE COMMENCER LA MANIPULATION**

1. Aller sur https://github.com/ pour créer un compte personnel : c'est un processus classique que vous connaissez (confirmation du compte via une boîte mail etc...). Néanmoins certains aspects sont particuliers à github

>**ATTENTION :** 
* **Je ne vous laisse PAS le choix concernant le Username !!**  
* En informatique, bien souvent une minuscule et une majuscule ce n'est pas pareil. Faîtes attention !! [Plus d'informations](https://fr.wikipedia.org/wiki/Sensibilit%C3%A9_%C3%A0_la_casse)

* Username : `prenomNom` c'est-à-dire respectant la notation [**lowerCamelCase**](https://fr.wikipedia.org/wiki/Camel_case) (*Evidemment il s'agit d'adaper `prenomNom` à votre identité personnelle*)

* Mot de passe : L'expérience des années précédentes montre que certains élèves ne se souviennent pas de leur mot de passe (certainement ceux qui n'ouvrent pas souvent leur github pour travailler...).

# Sécurité sur github

* **Ne stocker aucun mot de passe, aucune information personnelle sur github !!**


* Votre espace de stockage sur github est limité (mais il est bien suffisant pour y stocker tout votre travail) : inutile de tenter d'y ajouter de gros fichiers comme des vidéos ! **De toute façon, tout ce qui est stocké sur github doit concerner uniquement le travail en NSI. Tout autre usage sera sanctionné !**


* Vous n'êtes pas à l'abri d'une fausse manipulation sur votre _repository_ travail. Ainsi un élève a déjà perdu tout son travail réalisé en **projet** (donc nécessaire pour son oral de bac) suite à une fausse manipulation **de SA faute**. Il convient donc de réaliser **régulièrement** des sauvegardes de son _repository_ en local sur sa machine **à la maison**, par exemple toutes les 2 semaines. Ceci se fait en le téléchargeant grâce au bouton **clone or download** puis en sélectionnant **Download zip**. C'est une procédure simple qui prend quelques secondes.... cela vaut le coup plutôt que de perdre plusieurs heures de travail ! D'ailleurs tous les informaticiens font des sauvegardes locales de leur travail sur github, ne serait-ce que pour pouvoir continuer à travailler même en cas de panne réseau...   
*Remarque : il existe d'autres moyens plus efficaces et automatisées de faire des sauvegardes régulières d'un repository github, mais cela met en oeuvre des connaissances sur la technologie [git](https://fr.wikipedia.org/wiki/Git) que je ne vais pas aborder ici*


**Organiser correctement les documents informatiques est FONDAMENTAL en informatique**.
