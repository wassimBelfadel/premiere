# TP Construction d'une page HTML
(*D'après [FIL](https://www.fil.univ-lille1.fr/)*)

> * L'objet de cet exercice est de construire un document HTML. Comme il n'est pas question de CSS ici, la mise en forme est sans importance : ce qui importe c'est la structure du document (la mise en forme sera donc celle générée par défaut par le navigateur).  
* Certaines balises vous sont inconnues : votre travail consiste, **comme en TD**, de trouver l'information notamment sur le site du [Mozilla Development Network ou MDN](https://developer.mozilla.org/fr/). Ainsi c'est aussi votre capacité d'autonomie pour trouver et vous approprier de l'information technique qui est sollicitée ici (capacité très utile pour les informaticiens)  
* Il s'agit ici de se mettre dans la peau d'un [intégrateur web](https://fr.wikipedia.org/wiki/Int%C3%A9grateur_web) : il faut donc juste faire ce qui est demandé, surtout pas d'apporter votre touche personnelle !!

### Structure HTML

```html
<!DOCTYPE html>
<html lang=fr dir="ltr">
  <head>
    <meta charset="utf-8"/>
    <title></title>
  </head>
  <body>
    
  </body>
</html>
```


**A partir de :**
* **la structure HTML ci-dessus**
* **le texte brut fourni dans tim_berners-lee.txt**
* **le document tim_berners-lee.pdf**

**Construire le document HTML reprenant globalement la structure du document `tim_berners-lee.pdf` en respectant les règles suivantes  :**
1. Créez un entête approprié afin que :
    * Le titre de la page web soit Tim Berners Lee
    * L'encodage du texte soit en UTF-8 (métadonnée)
    * Vous soyez l'auteur de cette page (métadonnée)
    
    
2. Aux titres de section du PDF correspondront des sections de titre de même "niveau".


3. Les paragraphes de texte seront placés dans des éléments `<p>`


4. Utilisez les éléments de phrasé suivants :
    * pour les citations du texte, l'élément `<q>`
    * pour les textes mis en évidence en gras, l'élément `<strong>`,
    * pour le terme World Wide Web qui apparaît en légende de l'image de l'ordinateur NEXT, un élément `<em>`.
    
    
5. Vous placerez les images avec leur légende au sein d'un élément `<figure>`. Deux ont pour source un fichier local l'autre une cible distante. Chaque image doit avoir un un **texte alternatif** (une très courte description de la photo) qui s'affiche à la place de l'image dans le cas où celle-ci n'apparaîtrait pas)


6. Vous marquerez comme des abréviations les termes : CERN, MIT, W3C, URL, HTTP et HTML et préciserez leur attribut title. Ceci aura pour effet de créer une "infobulle" au survol de la souris :

![info bulle](img/infoBulle.png)
*(L'infobulle "Conseil Européen pour la recherche Nucléaire" apparaît lorsque la souris survole le mot "CERN")*


7. Les 3 notes de bas de page seront remplacées par des hyperliens qui porteront sur les mots suivis du numéro de la note de bas de page, la cible de cet hyperlien est le contenu de la note. Les numéros des notes ne doivent donc plus apparaitre dans le document HTML final.


8. Utilisez un moteur de recherche pour trouver une balise permettant de placer une portion de texte en "exposant", comme dans le texte "XX$^e $siècle" à la fin du document.


9. Créez des listes à puces (non ordonnées) pour les deux parties du texte avec des énumérations


10. Les dates du document seront identifiées par un élément `<time>`.


11. Vérifiez la [validité HTML](https://validator.w3.org/nu/) de votre document, et si besoin corrigez-le jusqu'à ce qu'il soit valide.


12. **RAPPEL : Tous les fichiers utiles pour votre page web doivent être placés dans un dossier `TP_HTML` déposé _au bon endroit_ dans votre _repository_ sur github _(Comme d'habitude en NSI)_**
